# Smart Finance Assistant: Budget Buddy

## Project Overview

This project creates a Smart Finance Assistant called Budget Buddy. The assistant helps users understand their personal finance situation by analysing transaction data from a CSV file.

The assistant can:
- Upload and read a transaction CSV file
- Summarise total income, expenses and net balance
- Show spending by category
- Answer finance-related questions based on uploaded data
- Calculate savings goals using a custom tool
- Present everything in a simple Gradio interface

- Calculate income and expenses
- Identify major spending categories

This project solves the problem by creating a Smart Finance Assistant that turns transaction data into useful finance insights.

```python
REQUIRED_COLUMNS = ["date", "description", "category", "amount", "type"]

def validate_csv(df):
    """
    Check whether the uploaded CSV has the required columns and is not empty.
    """
    missing_columns = [col for col in REQUIRED_COLUMNS if col not in df.columns]

    if missing_columns:
        return False, f"Missing required columns: {missing_columns}"

    if df.empty:
        return False, "The CSV file is empty."

    return True, "CSV file is valid."
```

```python
def load_transactions(file):
    """
    Load the uploaded CSV file and clean the transaction data.
    """
    try:
        if file is None:
            return None, "Please upload a CSV file."

        df = pd.read_csv(file.name)

        is_valid, message = validate_csv(df)
        if not is_valid:
            return None, message

        df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
        df["type"] = df["type"].astype(str).str.lower().str.strip()
        df["category"] = df["category"].astype(str).str.strip()
        df["description"] = df["description"].astype(str).str.strip()

        df = df.dropna(subset=["amount"])

        return df, "CSV loaded successfully."

    except Exception as e:
        return None, f"Error loading CSV: {str(e)}"
```

```python
def summarise_transactions(df):
    """
    Summarise income, expenses, net balance and spending by category.
    """
    if df is None or df.empty:
        return "No transaction data available."

    income = df[df["type"] == "income"]["amount"].sum()
    expenses = df[df["type"] == "expense"]["amount"].sum()
    balance = income - expenses

    expense_df = df[df["type"] == "expense"]

    category_spending = (
        expense_df
        .groupby("category")["amount"]
        .sum()
        .sort_values(ascending=False)
    )

    summary = f"""
## Finance Summary

Total income: ${income:,.2f}

Total expenses: ${expenses:,.2f}

Net balance: ${balance:,.2f}

## Spending by Category
"""

    if len(category_spending) == 0:
        summary += "\nNo expense categories found.\n"
    else:
        for category, amount in category_spending.items():
            summary += f"- {category}: ${amount:,.2f}\n"

    if income == 0 and expenses > 0:
        summary += "\nWarning: No income was found in the data. Please check your CSV file."
    elif expenses > income:
        summary += "\nWarning: Your expenses are higher than your income."
    elif balance < income * 0.1:
        summary += "\nWarning: Your remaining balance is less than 10% of your income."
    else:
        summary += "\nGood job: Your income is higher than your expenses."

    return summary
```

## RAG Component

This project uses a simple RAG-style approach. After the user uploads a CSV file, the program retrieves relevant information from the transaction data, such as total income, total expenses, net balance, top spending categories and recent transactions.

The chatbot then uses this retrieved context to answer finance-related questions based on the user's own data.

```python
def create_rag_context(df):
    """
    Create a simple RAG-style context from the uploaded CSV file.
    The chatbot uses this context to answer questions based on user data.
    """
    if df is None or df.empty:
        return "No data available."

    income = df[df["type"] == "income"]["amount"].sum()
    expenses = df[df["type"] == "expense"]["amount"].sum()
    balance = income - expenses

    top_categories = (
        df[df["type"] == "expense"]
        .groupby("category")["amount"]
        .sum()
        .sort_values(ascending=False)
        .head(5)
    )

    recent_transactions = df.tail(5)

    context = f"""
Transaction Data Context

Total income: {income}
Total expenses: {expenses}
Net balance: {balance}

Top spending categories:
"""

    if len(top_categories) == 0:
        context += "No expense categories found.\n"
    else:
        for category, amount in top_categories.items():
            context += f"- {category}: {amount}\n"

    context += "\nRecent transactions:\n"
    context += recent_transactions.to_string(index=False)

    return context
```

```python
def finance_chatbot(question, df):
    """
    Finance chatbot that answers questions based on uploaded transaction data.
    """
    if question is None or question.strip() == "":
        return "Please ask a finance-related question."

    if df is None or df.empty:
        return "Please upload a transaction CSV first so I can analyse your data."

    question_lower = question.lower()
    rag_context = create_rag_context(df)

    income = df[df["type"] == "income"]["amount"].sum()
    expenses = df[df["type"] == "expense"]["amount"].sum()
    balance = income - expenses

    category_spending = (
        df[df["type"] == "expense"]
        .groupby("category")["amount"]
        .sum()
        .sort_values(ascending=False)
    )

    if "summary" in question_lower or "summarise" in question_lower or "overview" in question_lower:
        return summarise_transactions(df)

    elif "highest" in question_lower or "most" in question_lower or "top" in question_lower:
        if len(category_spending) == 0:
            return "I could not find any expense categories in your data."

        highest_category = category_spending.index[0]
        highest_amount = category_spending.iloc[0]

        return f"""
Your highest spending category is **{highest_category}** with **${highest_amount:,.2f}**.

This means {highest_category} is the area where you may have the biggest opportunity to reduce spending.
"""

    elif "save" in question_lower or "saving" in question_lower:
        if balance <= 0:
            return """
You currently do not have a positive balance to save.

My advice:
- Review your highest spending category first
- Reduce non-essential expenses
- Try to create a small emergency fund when your balance becomes positive
"""
        else:
            suggested_saving = balance * 0.2
            return f"""
Your net balance is **${balance:,.2f}**.

A simple saving strategy is to save around 20% of your remaining balance.

Suggested saving amount: **${suggested_saving:,.2f}**

You can adjust this depending on your rent, bills and personal needs.
"""

    elif "advice" in question_lower or "recommend" in question_lower or "budget" in question_lower:
        if expenses > income:
            return """
Your expenses are higher than your income.

Budget advice:
- Reduce non-essential spending first
- Check your largest expense category
- Avoid taking on extra debt
- Set a small weekly spending limit
"""
        elif balance < income * 0.1:
            return """
Your budget is positive, but your remaining balance is low.

Budget advice:
- Try to reduce small daily expenses
- Set a fixed savings goal
- Track spending weekly
- Keep emergency savings separate
"""
        else:
            return """
Your budget is in a healthy position because your income is higher than your expenses.

Budget advice:
- Continue tracking spending
- Save part of your balance every month
- Review your largest spending category
- Build an emergency fund
"""

    elif "context" in question_lower or "data" in question_lower:
        return rag_context

    else:
        return f"""
I am your finance coach. Based on your uploaded transaction data:

- Total income: ${income:,.2f}
- Total expenses: ${expenses:,.2f}
- Net balance: ${balance:,.2f}

You asked: "{question}"

Try asking one of these:
- Give me a summary
- What is my highest spending category?
- Give me saving advice
- Give me budget advice
"""
```

## Custom Agent Tool: Savings Goal Calculator

The custom tool calculates how many months a user needs to reach a savings goal.

Inputs:
- Goal amount
- Current savings
- Monthly saving amount

Output:
- Remaining amount
- Estimated number of months needed

```python
def savings_goal_calculator(goal_amount, current_savings, monthly_saving):
    """
    Custom tool that calculates how many months are needed to reach a savings goal.
    """
    try:
        goal_amount = float(goal_amount)
        current_savings = float(current_savings)
        monthly_saving = float(monthly_saving)

        if goal_amount <= 0:
            return "Goal amount must be greater than zero."

        if current_savings < 0:
            return "Current savings cannot be negative."

        if monthly_saving <= 0:
            return "Monthly saving must be greater than zero."

        remaining = goal_amount - current_savings

        if remaining <= 0:
            return "Congratulations! You have already reached your savings goal."

        months_needed = math.ceil(remaining / monthly_saving)

        return f"""
## Savings Goal Result

Goal amount: ${goal_amount:,.2f}

Current savings: ${current_savings:,.2f}

Remaining amount: ${remaining:,.2f}

Monthly saving: ${monthly_saving:,.2f}

You need approximately **{months_needed} month(s)** to reach your goal.
"""

    except ValueError:
        return "Please enter valid numbers."
```

## Gradio User Interface

The Gradio interface connects all parts of the project. It includes three tabs:

1. Upload CSV and Summary
2. Finance Chatbot
3. Savings Goal Tool

```python
stored_df = None

def upload_and_summarise(file):
    """
    Upload CSV, store the dataframe and return a finance summary.
    """
    global stored_df

    stored_df, message = load_transactions(file)

    if stored_df is None:
        return message

    return message + "\n\n" + summarise_transactions(stored_df)

def chat_interface(question):
    """
    Connect chatbot to the stored uploaded dataframe.
    """
    global stored_df
    return finance_chatbot(question, stored_df)

with gr.Blocks(title="Smart Finance Assistant") as demo:
    gr.Markdown("# Smart Finance Assistant: Budget Buddy")
    gr.Markdown(
        "Upload your transaction CSV, ask finance questions, and calculate savings goals."
    )

    with gr.Tab("1. Upload CSV and Summary"):
        csv_file = gr.File(label="Upload transaction CSV")
        summary_output = gr.Markdown()
        upload_button = gr.Button("Analyse Transactions")

        upload_button.click(
            upload_and_summarise,
            inputs=csv_file,
            outputs=summary_output
        )

    with gr.Tab("2. Finance Chatbot"):
        question_input = gr.Textbox(
            label="Ask a finance question",
            placeholder="Example: What is my highest spending category?"
        )
        chatbot_output = gr.Markdown()
        chat_button = gr.Button("Ask Assistant")

        chat_button.click(
            chat_interface,
            inputs=question_input,
            outputs=chatbot_output
        )

    with gr.Tab("3. Savings Goal Tool"):
        goal_input = gr.Number(label="Goal amount")
        current_input = gr.Number(label="Current savings")
        monthly_input = gr.Number(label="Monthly saving")
        savings_output = gr.Markdown()
        savings_button = gr.Button("Calculate Savings Goal")

        savings_button.click(
            savings_goal_calculator,
            inputs=[goal_input, current_input, monthly_input],
            outputs=savings_output
        )

demo.launch(share=True, debug=True)
```
